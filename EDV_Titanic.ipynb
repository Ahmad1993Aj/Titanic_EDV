```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic.csv")
df.head()
```

```python
df.drop(columns="PassengerId", inplace = True)
```

```python
df.head()
```

```python
df.drop(columns="Ticket", inplace = True)
```

```python
df.rename(columns={
    "Pclass":"Passenger_class",
    "Anzahl_Geschwister_Ehegatten_an_Bord":"Geschwister_Ehegatten",
    "Anzahl_Eltern_Kinder_an_Bord":"Eltern_Kinder"
}, inplace =True)
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df,
            x=df["Age"],
            kde=True,
            bins=20)
plt.title("Distrubtion of Age",
          fontsize=18)
plt.xlabel("Age",
          fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df,
            x=df["Passenger_class"],
            kde=True,
            bins=20)
plt.title("Distrubtion of Age",
          fontsize=18)
plt.xlabel("Age",
          fontsize=12)
plt.show()
```

```python
for col in df.columns:
    print(f"============== {col} ===================")
    if df[col].dtype != "object":
        plt.figure(figsize=(12,6))
        sns.histplot(data=df,
                    x=df[col],
                    kde=True,
                    bins=20)
        plt.title(f"Distrubtion of {col}",
                  fontsize=18)
        plt.xlabel(f"{col}",
                  fontsize=12)
        plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,
           x="Survived",
           y="Age")
plt.title("Age vs Survived",
          fontsize=18)
plt.xlabel("Survived", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.show()
```

```python
for col in df.columns:
    print(f"============== {col} ===================")
    if df[col].dtype != "object":
        plt.figure(figsize=(12,6))
        sns.boxplot(data=df,
           x="Survived",
           y="Age")
        plt.title(f"Distrubtion of {col}",
                  fontsize=18)
        plt.xlabel(f"Survived",
                  fontsize=12)
        plt.ylabel(f"{col}",
                  fontsize=12)
        plt.show()
```
